splitting our dataset into train and test and then we will remove skewness because i want my test data untouchable

```python
df_x=df.drop(columns=['W'])
y=df[['W']]
```

```python
for i in df_x.columns:
    if df_x[i].skew()>0.5:
        df_x[i]=np.cbrt(df_x[i])
    if df_x[i].skew()<-0.5:
        df_x[i]=np.cbrt(df_x[i])
        
df_x.skew()
```

```python
df_x.head()
```

Model building and finding best random state

```python
lr=LinearRegression()
from sklearn.metrics import r2_score
```

```python
for i in range(0,100):
    df_x_train,df_x_test,y_train,y_test=train_test_split(df_x,y,test_size=0.2,random_state=i)
    lr.fit(df_x_train,y_train)
    pred_train = lr.predict(df_x_train)
    pred_test = lr.predict(df_x_test)
    print("At Random State",i,"the traning accuracy is :-", r2_score(y_train,pred_train))
    print("\n")
```

```python
df_x_train,df_x_test,y_train,y_test=train_test_split(df_x,y,test_size=0.2,random_state=2)
```

```python
lr.fit(df_x_train,y_train)
pred_test = lr.predict(df_x_test)
print(r2_score(y_test,pred_test))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
rf.fit(df_x_train,y_train.values.ravel())
predrf = rf.predict(df_x_test)

rfs = r2_score(y_test,predrf)
print('R2 Score',rfs*100)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
dt.fit(df_x_train,y_train)
preddt = dt.predict(df_x_test)

dts = r2_score(y_test,preddt)
print("dt score",dts*100)
```
